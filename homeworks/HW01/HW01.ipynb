# Импорт библиотеки NumPy

```python
import numpy as np
```

# Создание массива из списка

```python
a = np.array([1,2,3,4,5])
b = np.array(range(1,7)) # альтернативный способ создания массива
print(a)
print(type(a))
print(np.shape(a))
print(b)
print(type(b))
print(np.shape(b))
```

# Создание матрицы нулей 

```python
a = np.zeros((5,5))
print(a)
print(type(a))
print(np.shape(a))
```

# Создание массива единиц

```python
a = np.ones((4,3)) 
print(a)
print(type(a))
print(np.shape(a))
```

# Создание единичной матрицы

```python
a = np.eye(5,5) 
print(a)
print(type(a))
print(np.shape(a))
```

# Создание незаполненного (заполненного мусором) массива

```python
a = np.empty((3,3,3)) 
print(a)
print(type(a))
print(np.shape(a))
```

# Создание массива из последовательности чисел numpy

```python
a = np.arange(9,0,-1.5) # создание последовательности
print(a)
b = np.array([a,a]) # создание двумерного массива из последовательности
print(b)
```

# Создание массива, заполненного inf

```python
a = np.full((2,2), np.inf) 
print(a)
print(type(a))
print(np.shape(a))
```

# Демонстрация аттрибутов для массива

```python
a = np.full((2,2), 15) # заполненый числом 15 массив
print(a)
print(type(a))
print(np.shape(a))
print(a.dtype)
```

# Примеры индексирования и срезов

```python
a = np.array([2,1,3,6,5,3,8,71]) # одномерный массив
print(a)
print(a[4]) # получение отдельного элемента
print(a[1:4]) # получение подмассива элементов с помощью среза

```

```python
b = np.array([[2,1,3], 
            [6,10,5],
            [4,7,71]]) # двумерный массив
print(b)
print(b[2,1]) # второй столбец третья строка
print(b[[0,2]]) # выбор первой и третьей строки из матрицы
```

```python
print(b[0:2]) # срез с первой по вторую строку матрицы
print(b[0:2,0:2]) # срез с первой строки по вторую строку и с первого столбца по второй столбец
```

```python
b[2][2] # другой способ индексирования
```

```python
b[:,2] # весь третий столбец
```

```python
b[:2] # первая и вторая строка
```

```python
b[...,2] # то же самое, что и [:,2]
```

# Примеры векторных операций и базовых статистик

```python
a = np.array([[1,2,3], 
            [4,6,8],
            [9,25,12]])
b = np.array([[1,9,2], 
            [1,6,8,],
            [1,0,1]])
```

```python
print(a + b) # поэлементное сложение массивов
```

```python
print(a * b) # поэлементное умножение массивов
```

```python
print(a ** b) # поэлементное возведение массива a в степень b (элементов a в степень элементов b)
```

```python
print(a.sum()) # сумма элементов массива a
```

```python
print(b.mean()) # вычисление средней величины массива b
```

```python
print(b.std()) # вычисление стандартного отклонения массива b
```
